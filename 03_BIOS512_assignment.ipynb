Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.height=10,repr.plot.width=11)

ggplot(penguins,aes(x=body_mass_g)) + geom_histogram(bins=34, fill='lightblue', color="#7bc4e6") +
scale_x_continuous(labels=scales::comma_format()) +
facet_wrap(~year, ncol=1) +
labs(x="Body Mass(g)", y="Number of penguins", title="Histograms for penguin body mass, 2007-2009") + 
theme_minimal() + 
theme(plot.title = element_text(face="bold", size=18))
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
options(repr.plot.height=10,repr.plot.width=10)
ggplot(penguins, aes(x=body_mass_g, y=sex, color=sex)) + 
    geom_jitter(height=0.11, alpha=0.4, size=3) +
    scale_x_continuous(labels=scales::comma_format()) +
    facet_wrap(~species, ncol=1) +
    labs(x="Body Mass(g)", 
        y="Sex", 
        title="Sex differences in body mass by penguin species") + 
    theme_minimal() + 
    theme(plot.title = element_text(face="bold", size=18),
     panel.background = element_rect(color="black"),
     plot.title.position = "plot",
     legend.position = "bottom"
    )
```
