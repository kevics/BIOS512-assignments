```r
library('tidyverse')

wind_turbine = readr::read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv')
# readcsv reads in data as tibble

# can print tibble, which only displays subset but tells you the dimensions
# print df would show ALL rows, while head() shows the subset without dimensions but it does format nicely
wind_turbine |> head()
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
wind_turbine |>
    mutate(province_territory = province_territory |> fct_infreq() |> fct_rev()) |>
    ggplot(aes(y=province_territory)) + geom_bar()


```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
wind_turbine |>
    mutate(manufacturer = manufacturer |> 
                            fct_infreq() |> fct_rev() |>
                            fct_lump_n(n = 6, other_level = "Other") |>
                            fct_relevel("Other", after = 0)
                            ) |>
    ggplot(aes(y=manufacturer)) + geom_bar()
```

```r
wind_turbine |>
    group_by(manufacturer) |>
    count() |>
    arrange(n)

print(paste0("# manufacturers with more than 100 = ", 
        wind_turbine |>
        group_by(manufacturer) |>
        count() |>
        filter(n > 100) |>
        nrow()
        )
)
```
